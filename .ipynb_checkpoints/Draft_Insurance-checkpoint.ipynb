# Introduction


```python
# Packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

# Utils

# EDA

```python
df = pd.read_csv("insurance-adapted.csv")
```

```python
# Get x
age = df.age
```

```python
# Get Y
charges = df.charges
```

```python
smoker = df.smoker

label_encoder = LabelEncoder()

df['smoker_encoded'] = label_encoder.fit_transform(df['smoker'])
df = df.drop(columns=['smoker'])
```

```python
#plt.scatter(age, charges, color="red")

categories = np.array(df['smoker_encoded'])

# Smokers are in Green, Non-Smokers are in red
colormap = np.array(['r', 'g'])

plt.scatter(age, charges, c=colormap[categories])


plt.title("Check relation betweeen x and y")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
